```python
import os,sys,inspect
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir) + "\\Bahamas"
sys.path.insert(0,parentdir) 
parentdir

import optionchain
cManager  = optionchain.OptionChainManager()
cManager.addNewInstrument("RUT")
```

```python
import spread
spreadC = spread.SpreadConstructor(cManager)
newSpread = spreadC.createWidthValueSpread("RUT","19/02/2016","Credit","Put",30,2.7,"04/01/2016")

dates = list(newSpread.contractA.marketData.keys())
dates.sort()
prices = []
for key in dates:
    prices.append((newSpread.contractA.marketData[key].ask - newSpread.contractB.marketData[key].ask) * 100)

df = pd.DataFrame({'date' : dates, 'spread price' : prices})
df.set_index('date', inplace=True)    
ax = df.plot(figsize=(10,5), grid=True)
print("Selected strikes " + newSpread.contractA.strike + " and " + newSpread.contractB.strike)
```
