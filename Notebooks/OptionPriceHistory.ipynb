```python
import optionchain
import pandas as pd
```

```python
cManager  = optionchain.OptionChainManager()
cManager.addNewInstrument("RUT")
```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

def f(strike_price):
    tryGetContract = cManager.getContractData("RUT","19/02/2016","Call",str(strike_price))
    
    if tryGetContract is None:
        print("Contract not found")
        pass
    
    dates = list(tryGetContract.marketData.keys())
    dates.sort()
    prices = []
    for key in dates:
        prices.append(tryGetContract.marketData[key].ask)

    df = pd.DataFrame({'date' : dates, 'mid' : prices})
    df.set_index('date', inplace=True)    
    ax = df.plot(figsize=(10,5), grid=True)
    
    
interact(f,strike_price=widgets.IntSlider(min=800,max=1200,step=5,value=1000))
```
